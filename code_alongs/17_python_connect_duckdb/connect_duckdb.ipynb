## Connect duckdb in python

```python
import duckdb

with duckdb.connect("database/jokes.duckdb") as conn:
    conn.sql("DROP TABLE IF EXISTS jokes;")
    
    conn.sql("""--sql
        CREATE TABLE IF NOT EXISTS jokes ( 
            id INT, 
            joke STRING
        );
    """)

    conn.sql("""--sql
        INSERT INTO jokes VALUES
            (1, 'Why do programmers hate sunllight, becuase they attract bugs'),
            (2, 'Why do java programmers need glasses, becuase they cannot C#');
        """)

    desc = conn.sql("DESC;").df()
```

```python
with duckdb.connect("database/jokes.duckdb") as conn:
    jokes = conn.sql("FROM jokes;").df()

# outside -> db connection is closed
# but have saved dataframe to jokes variable
jokes
```

## Ingest a csv file into duckdb

- we can open a csv file without a database connection

```python
# same as pd.read_csv()
finance = duckdb.sql("""
    FROM 'data/financial_data.csv'
""").df()

finance.head(2)
```

```python
finance.info()
```

```python
finance["category"].value_counts()
```

```python
df_category_amount = duckdb.sql("""
    SELECT category, SUM(amount) AS total_amount
    FROM finance
    GROUP BY category
    ORDER BY total_amount DESC;
""").df()
df_category_amount
```

```python
df_category_amount.plot(
    x="category", y="total_amount", kind="bar", title="Total Amount by Category", xlabel='Category', ylabel='Total Amount',
)
```
