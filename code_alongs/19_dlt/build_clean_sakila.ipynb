```python
import duckdb

con = duckdb.connect("data/clean_sakila.duckdb")

tables = [
    "film", "actor", "category", "film_actor", "film_category",
    "language", "store", "staff", "customer", "rental",
    "inventory", "address", "city", "country"
]

for t in tables:
    con.execute(f"""
        CREATE TABLE {t} AS 
        SELECT * FROM read_csv_auto('sakila_csv/{t}.csv')
    """)

con.close()
print("DONE")
```
